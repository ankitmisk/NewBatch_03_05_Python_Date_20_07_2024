```python
import pygame
pygame.init()
import time
from gtts import gTTS
import IPython
question_bank = ['In what year did the Great October Socialist Revolution take place?',
                'What is the largest lake in the world?',
                 'Which planet in the solar system is known as the Red Planet?',
                'Who wrote the novel War and Peace?',
                 'What is the capital of Japan?']

options = [['1917', '1923', '1914', '1920'],
          ['Caspian Sea', 'Baikal', 'Lake Superior', 'Ontario'],
          ['Venus', 'Earth', 'Mars', 'Jupiter'],
          ['Anton Chekhov', 'Fyodor Dostoevsky', 'Leo Tolstoy', 'Ivan Turgenev'],
          ['Beijing', 'Tokyo', 'Seoul', 'Bangkok']]

correct_op = ['A','B','C','C','B']


op_list =  'ABCD'
money = 0


# nested loop

print('Welcome to KBC'.center(120,' '),end = '\n'*3)


file =  r"C:\Users\Lenovo\Downloads\kbc-awesome-5410 (2).mp3"

kbc = pygame.mixer.Sound(file)
kbc.play()




time.sleep(5)

kbc.stop()

for index,value in enumerate(question_bank):
    print(f"Q{index+1}.{value}")
    
    data = value
    audio = gTTS(data)
    audio.save('question.mp3')
    
    mera_sound = pygame.mixer.Sound('question.mp3')
    mera_sound.play()
    
    
    
    
    for op_index,op_value in enumerate(options[index]):
        print(f" {op_list[op_index]}. {op_value}")
    
    
    ans = input('Enter your choice: ')
    
    if ans.lower() in 'abcd':
        if ans.upper() == correct_op[index]:
            if index == len(question_bank) -1:
                final_win_audio = r"C:\Users\Lenovo\Downloads\7-crore-kbc-sound-effect-made-with-Voicemod.mp3"
                
                sound = pygame.mixer.Sound(final_win_audio)
                sound.play()
                display(IPython.display.Image(r'https://media.tenor.com/E6WSW38Kn94AAAAM/kbc-great.gif'))                
            else:
                print('You won!!')

                correct = r"C:\Users\Lenovo\Downloads\correct-6033.mp3"

                sound = pygame.mixer.Sound(correct)
                sound.play()

            
            money += 100000
        else:
            print('You Lose!!')
            wrong = r"C:\Users\Lenovo\Downloads\_Kbc_Wrong_Answer_Sound_Ringtone_(by Fringster.com).mp3"
            
            sound = pygame.mixer.Sound(wrong)
            sound.play()
            
            print('Bye!!Bye!!')
            break
        
        
    else:
        print('Invalid op/choice')
        print('No attempt left!!')
        break
        
    print(end = '\n'*3)
    
    display(clear=True)
    
        
   
```
